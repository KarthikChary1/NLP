```python
import nltk
stem=nltk.stem.PorterStemmer()
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
sentence
```

```python
data_text=[]
for i in range(len(data.tweet_text)):
    word=re.sub('[^a-zA-Z]'," ",str(data.tweet_text[i]))
    words = word.lower()
    word=word.split()
    word = [stem.stem(i) for i in word if i not in set(stopwords)]
    sentence=" ".join(word)
    data_text.append(sentence)
```

```python
word_token=[nltk.word_tokenize(i) for i in data_text]
```

```python
li=[]
for i in word_token:
    for j in i:
        li.append(j)
```

```python
fd=nltk.FreqDist(li)
common=fd.most_common(20)
```

```python
common_list=[]
for i,v in common:
    common_list.append(i)
```

```python
final_list=[]
for i in word_token:
    sen=[word for word in i if word in common_list]
    sen=" ".join(sen)
    final_list.append(sen)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
```

```python
df=pd.DataFrame(vectorizer.fit_transform(final_list).toarray(),columns=vectorizer.get_feature_names())
```
